```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
#Create new list for processed_track_lyrics 
#Remove some stop words from spacy's default list (mostly pronouns and contractions)
#Cut singing sounds 
#Keep question mark and apostraphes but cut dashes and commas 
#make everything lowercase 

processed_track_lyrics = []

not_stop_words = ["i", 
                      "'m", 
                      "she", 
                      "'s", 
                      "he", 
                      "'s",
                      "that",
                      "you", 
                      "me", 
                      "us", 
                      "we", 
                      "'re", 
                      "our", 
                      "ours", 
                      "it", 
                      "'s",
                      "they", 
                      "'re"]

for word in not_stop_words:
        nlp.vocab[word].is_stop = False

sing_sounds = ["ooh",
               "oh",
               "ohohohohoh",
               "ohohohohoh",
               "mmmmm",
               "mm",
               "mmm",
               "hey",
               "heyyyy",
               "la",
               "yeah"]
               
import string

punctuation = string.punctuation.replace("?", "").replace("'", "").replace("'","").replace("'", "").replace("[","").replace("]","")

for index in range(len(chris_cornell_df)):
    text = chris_cornell_df.loc[index, "track_lyrics"].lower().strip()
    text = " ".join(text.split())
    
    non_punctuation = "".join(ch for ch in text if ch not in punctuation)
    
    doc=nlp(non_punctuation)
    
    processed_text = " ".join(token.text for token in doc if not token.is_stop and token.text not in sing_sounds)
    processed_text = processed_text.replace(" '", "'")
    processed_text = processed_text.replace("wo n't", "won't")
    processed_text = processed_text.replace("ca n't", "can't")
    processed_text = processed_text.replace("do n't", "don't")
    processed_text = processed_text.replace("does n't", "doesn't")
    processed_text = processed_text.replace("did n't", "didn't")
    processed_text = processed_text.replace("for 's", "for's")
    processed_track_lyrics.append(processed_text)

#https://www.geeksforgeeks.org/nlp/stop-words-using-spacy/
#https://stackoverflow.com/questions/6142915/remove-unwanted-space-in-between-a-string
#https://www.geeksforgeeks.org/python/string-punctuation-in-python/
#https://stackoverflow.com/questions/41170726/add-remove-custom-stop-words-with-spacy
```
