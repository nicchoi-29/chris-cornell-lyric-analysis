```python
#Loop for lexical diversity value for each song in 'processed_track_lyrics' column 

song_word_diversity_df = chris_cornell_df_without_labels[["album_name", "track_title", "processed_track_lyrics", "release_year", "track_lyrics"]].copy()
```

```python
#Create new lists to add as new columns to new dataframe 

word_diversity_score = []
word_count_processed_lyrics = []
unique_text_count_processed_lyrics = []

text = song_word_diversity_df['processed_track_lyrics']

for text in song_word_diversity_df["processed_track_lyrics"]:
    tokens = text.split()
    def lexical_diversity(text):
        if len(text) == 0:
            return 0
        return len(set(text)) / len(text)
    word_diversity_score.append(lexical_diversity(tokens) * 100)
    word_count_processed_lyrics.append(len(tokens))
    unique_text_count_processed_lyrics.append(len(set(tokens)))

    
song_word_diversity_df["word_diversity_score"] = word_diversity_score
song_word_diversity_df["unique_text_count_processed_lyrics"] = unique_text_count_processed_lyrics
song_word_diversity_df["word_count_processed_lyrics"] = word_count_processed_lyrics

song_word_diversity_df = song_word_diversity_df[
    ["album_name", "release_year", "track_title", "word_diversity_score", "unique_text_count_processed_lyrics", "word_count_processed_lyrics", "processed_track_lyrics", "track_lyrics"]
]
```

6. By how much did word diversity change from first album to last album for solo, Soundgarden, and Audioslave?

```python
album_word_diversity_df = song_word_diversity_df.copy()
```

```python
album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Euphoria Mourning', 'unique_text_count_processed_lyrics'].sum()
```

```python
#calculate word diveristy mean for each album 
#use chart to show change 

unique_word_count_euphoriamourning = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Euphoria Mourning', 'unique_text_count_processed_lyrics'].sum()

unique_word_count_highertruth = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Higher Truth', 'unique_text_count_processed_lyrics'].sum()

unique_word_count_ultramegaok = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Ultramega OK', 'unique_text_count_processed_lyrics'].sum()

unique_word_count_kinganimal = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'King Animal', 'unique_text_count_processed_lyrics'].sum()

unique_word_count_audioslave = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Audioslave', 'unique_text_count_processed_lyrics'].sum()

unique_word_count_revelations = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Revelations', 'unique_text_count_processed_lyrics'].sum()
```

```python
total_word_count_euphoriamourning = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Euphoria Mourning', 'word_count_processed_lyrics'].sum()

total_word_count_highertruth = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Higher Truth', 'word_count_processed_lyrics'].sum()

total_word_count_ultramegaok = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Ultramega OK', 'word_count_processed_lyrics'].sum()

total_word_count_kinganimal = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'King Animal', 'word_count_processed_lyrics'].sum()

total_word_count_audioslave = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Audioslave', 'word_count_processed_lyrics'].sum()

total_word_count_revelations = album_word_diversity_df.loc[album_word_diversity_df['album_name'] == 'Revelations', 'word_count_processed_lyrics'].sum()
```

```python
word_diversity_score_euphoriamourning = unique_word_count_euphoriamourning/total_word_count_euphoriamourning*100

word_diversity_score_highertruth = unique_word_count_highertruth/total_word_count_highertruth*100

word_diversity_score_ultramegaok = unique_word_count_ultramegaok/total_word_count_ultramegaok*100

word_diversity_score_kinganimal = unique_word_count_kinganimal/total_word_count_kinganimal*100

word_diversity_score_audioslave = unique_word_count_audioslave/total_word_count_audioslave*100

word_diversity_score_revelations = unique_word_count_revelations/total_word_count_revelations*100
```

```python
album_word_diversity_df=album_word_diversity_df.drop_duplicates('album_name')
```

```python
album_word_diversity_df.drop(album_word_diversity_df[album_word_diversity_df["album_name"] == "Temple of the Dog"].index, inplace=True)
```

```python
album_word_diversity_df.shape
```

```python
album_word_diversity_df['average_album_word_diversity']= [word_diversity_score_euphoriamourning,
                                                          word_diversity_score_highertruth,
                                                          word_diversity_score_ultramegaok,
                                                          word_diversity_score_kinganimal,
                                                          word_diversity_score_audioslave,
                                                          word_diversity_score_revelations]
```

```python
album_word_diversity_df=album_word_diversity_df[['album_name','average_album_word_diversity']].sort_values('average_album_word_diversity',ascending = False)
```

```python
album_word_diversity_df.reset_index(drop=True, inplace = True)
```

```python
album_word_diversity_df
```

```python
album_word_diversity_df.loc[0:2, "first_album"] = ['Euphoria Mourning', 'Ultramega OK', 'Audioslave']
album_word_diversity_df.loc[0:2, "last_album"] = ['Higher Truth', 'King Animal', 'Revelations']

album_word_diversity_df["start_value"] = album_word_diversity_df["first_album"].map(
    album_word_diversity_df.set_index("album_name")["average_album_word_diversity"]
)

album_word_diversity_df["end_value"] = album_word_diversity_df["last_album"].map(
    album_word_diversity_df.set_index("album_name")["average_album_word_diversity"]
)
```
