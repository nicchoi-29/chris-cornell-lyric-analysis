```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# Source - https://stackoverflow.com/a/51627002
# Posted by Romain, modified by community. See post 'Timeline' for change history
# Retrieved 2026-07-10, License - CC BY-SA 4.0

print(nlp.Defaults.stop_words)
```

```python
#Create new list for processed_track_lyrics 
#Remove some stop words from spacy's default list (mostly pronouns and contractions)
#Cut singing sounds 
#Keep question mark and apostraphes but cut dashes and commas 
#make everything lowercase 

processed_track_lyrics =[]

import string

not_stop_words = ["i", 
                  "'m", 
                  "myself",
                  "she", 
                  "'s", 
                  "he", 
                  "'s",
                  "hers",
                  "her",
                  "his",
                  "that",
                  "you", 
                  "yourself",
                  "yourselves",
                  "herself",
                  "himself",
                  "me", 
                  "us", 
                  "we", 
                  "'re", 
                  "our", 
                  "ours", 
                  "they", 
                  "'re",
                  "won't",
                 "world",
                 "woman",
                 "man",
                 "ca",
                  "does",
                 "n't",
                 "was",
                 "do",
                 "did",
                 "would",
                 "was",
                 "ain't",
                 "it",
                 "never",
                 "not",
                 "none",
                 "no"]

added_stop_words = ["verse",
                "1",
                "2",
                "3",
                "4",
                "outro",
                "bridge",
                "chorus",
                "post",
                "pre",
                "chris",
                "cornell",
                "eddie",
                "vedder",
                "guitar",
                "solo",
                "hook",
                "vocal",
                "ad-libs",
                "interlude",
                "instrumental",
                "intro",
                "prechorus",
                "postchorus",
                "ad",
                "libs"]

sing_sounds = ["ooh",
           "oh",
           "ohohohohoh",
           "ohohohohoh",
           "mmmmm",
           "mm",
           "mmm",
           "hey",
           "heyyyy",
           "la",
           "yeah",
            "heyyyy-ya-la-la-hey-hey",
            "oh",
            "oh-oh-oh-oh-oh",
            "ooo",
            "oh-oh",
            "ooh-ooh-ooh-ooh",
            "ooh",
            "oohoohoohooh",
            "ohoh",
            "ohohoh",
            'ohooh',
            "oohooh",
            "uh",
            "huh",
            "heyyyyyalalaheyhey",
            "a-say",
            "a-",
            "adlibs",
            "asay",
            "ah",
            "yeahyeah",
            "oh-woah",
            "whoo",
            "oh-ooh-ooh-ooh-ooh",
            "ohoohoohoohooh",
            "ahahahohoh",
            "ah-ah-ah-oh-oh",
            "a-yeah",
            "ayeah",
            "huah"] 

punctuation = string.punctuation.replace("?", "").replace("'", "")

for word in not_stop_words:
     nlp.vocab[word].is_stop = False

for word in added_stop_words:
    nlp.vocab[word].is_stop = True

for index in range(len(chris_cornell_df_without_labels)):
    text = chris_cornell_df_without_labels.iloc[index]["track_lyrics"].lower().strip()
    text = " ".join(text.split())
    text = text.replace("adon't", "don't")
    text = text.replace("a-don't", "don't")
    text = text.replace('"think it\'s on d"', '')
    text = text.replace('i think so"', '')
    text = text.replace('"i just hit the wrong chord"', '')
    text = text.replace("…", "")
    non_punctuation = "".join(ch for ch in text if ch not in punctuation)

    doc=nlp(non_punctuation)

    processed_text = " ".join(token.text for token in doc if not token.is_stop and token.text not in sing_sounds)
    processed_text = processed_text.replace(" '", "'")
    processed_text = processed_text.replace("wo n't", "won't")
    processed_text = processed_text.replace("ca n't", "can't")
    processed_text = processed_text.replace("do n't", "don't")
    processed_text = processed_text.replace("did n't", "didn't")
    processed_text = processed_text.replace("for 's", "for's")
    processed_text = processed_text.replace("it 's", "it's")
    processed_text = processed_text.replace("was 't", "wasn't")
    processed_text = processed_text.replace("is n't", "isn't")
    processed_text = processed_text.replace("would n't", "wouldn't")
    processed_text = processed_text.replace("ai n't", "ain't")
    processed_text = processed_text.replace("got ta", "gotta")
    processed_text = processed_text.replace("gon na", "gonna")
    processed_text = processed_text.replace("afrom", "from")
    processed_text = processed_text.replace("adon't", "don't")
    processed_text = processed_text.replace("a-don't", "don't")
    processed_text = processed_text.replace("does n't", "doesn't")
    processed_text = processed_text.replace("uhbeautiful", "beautiful")

    processed_track_lyrics.append(processed_text)

#https://www.geeksforgeeks.org/nlp/stop-words-using-spacy/
#https://stackoverflow.com/questions/6142915/remove-unwanted-space-in-between-a-string
#https://www.geeksforgeeks.org/python/string-punctuation-in-python/
#https://stackoverflow.com/questions/41170726/add-remove-custom-stop-words-with-spacy
```
