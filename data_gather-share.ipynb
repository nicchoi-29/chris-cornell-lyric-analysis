```python
track_title
```

```python
euphoria_mourning_df['track_title'] = track_title
```

```python
euphoria_mourning_df
```

```python
track_lyrics = []

for lyric in euphoria_mourning["tracks"]:
    track_lyrics.append(lyric['song']['lyrics'])
```

```python
track_lyrics
```

```python
euphoria_mourning_df['track_lyrics'] = track_lyrics
```

```python
euphoria_mourning_df['album_name']= euphoria_mourning['name']
```

```python
euphoria_mourning_df
```
