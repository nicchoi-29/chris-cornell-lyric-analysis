```python
#Create new dataframe for pronoun counts per album and by each pronoun person (first, second, third)

album_data = chris_cornell_df_without_labels[
    chris_cornell_df_without_labels["album_name"].isin(
        ["Euphoria Mourning",
         "Higher Truth",
         "Ultramega OK",
         "King Animal",
         "Audioslave",
         "Revelations",
        "Temple of the Dog"])]

counts = []

for album, group in album_data.groupby("album_name"):
    album_words = " ".join(group["processed_track_lyrics"]).split()
    counter = Counter(album_words)
    
    counts.append({
        "album": album,
        "pronoun": "First",
        "pronoun_count": sum(counter[first_person] for first_person in first)
    })
    
    counts.append({
        "album": album,
        "pronoun": "Second",
        "pronoun_count": sum(counter[second_person] for second_person in second)
    })
    
    counts.append({
        "album": album,
        "pronoun": "Third",
        "pronoun_count": sum(counter[third_person] for third_person in third)
    })


pronoun_count_df = pd.DataFrame(counts)

pronoun_count_df
```

```python
word_count_totals = {
    'Euphoria Mourning': 1535,
    'Higher Truth': 2280,
    'Ultramega OK': 908,
    'King Animal': 1483,
    'Audioslave': 1569,
    'Revelations': 1545,
    'Temple of the Dog': 1375}

pronoun_count_df['total_word_count'] =  pronoun_count_df['album'].map(word_count_totals)
```

```python
pronoun_count_df['pronoun_percentage'] = (pronoun_count_df['pronoun_count'] / pronoun_count_df['total_word_count'])*100
```

```python
pronoun_count_df=pronoun_count_df.round(2)
```

```python
pronoun_bar_chart_df = pronoun_count_df.pivot(
    index="album",
    columns="pronoun",
    values="pronoun_percentage").reset_index()
```
